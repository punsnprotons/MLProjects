```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt


breast_cancer_data = load_breast_cancer()
print(breast_cancer_data.data[0])
print(breast_cancer_data.feature_names)

print(breast_cancer_data.target)
print(breast_cancer_data.target_names)

dataset = breast_cancer_data.data
labels = breast_cancer_data.target
#test_size = 0.2
#random_state = 100

training_data,validation_data,training_labels,validation_labels = train_test_split(dataset,labels,test_size=0.2,random_state=100)

print(len(training_data))
print(len(training_labels))

c_arr = []
k_arr = []
for k in range(1,101):
  k_arr.append(k)
  classifier = KNeighborsClassifier(n_neighbors =k)
  classifier.fit(training_data,training_labels)
  c_score = classifier.score(validation_data,validation_labels)
  c_arr.append(c_score)

c_max = max(c_arr)
idx_max = c_arr.index(c_max)
k_val_max = k_arr[idx_max]
print('Max validation accuracy:' + str(c_max))
print('K-value : ' + str(k_val_max))

plt.plot(k_arr,c_arr)
plt.xlabel('k')
plt.ylabel('Validation Accuracy')
plt.title('Breast Cancer Classifier Accuracy')
plt.show()
```
