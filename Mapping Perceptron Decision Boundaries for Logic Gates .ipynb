```python

from sklearn.linear_model import Perceptron
import matplotlib.pyplot as plt
import numpy as np
from itertools import product

# AND gate
and_data = [[0, 0], [0, 1], [1, 0], [1, 1]]
and_labels = [0, 0, 0, 1]

# Create Perceptron model for AND gate
classifier_and = Perceptron(max_iter=40, random_state=22)
classifier_and.fit(and_data, and_labels)

# Print the Perceptron score for AND gate
print('AND gate Perceptron Score:')
print(classifier_and.score(and_data, and_labels))

# Make predictions for specific points
results = classifier_and.decision_function([[0, 0], [1, 1], [0.5, 0.5]])
print(results)

# Generate a grid of points for visualization
x_values = np.linspace(0, 1, 100)
y_values = np.linspace(0, 1, 100)
point_grid = list(product(x_values, y_values))

# Calculate distances from decision boundary for each point
distances = classifier_and.decision_function(point_grid)
abs_distances = [abs(point) for point in distances]
distances_matrix = np.reshape(abs_distances, (100, 100))

# Plot the decision boundaries for AND gate
heatmap = plt.pcolormesh(x_values, y_values, distances_matrix)
plt.colorbar(heatmap)
plt.title('Perceptron decision boundaries for AND gate')
plt.show()
plt.clf()


```
